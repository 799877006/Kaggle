```python
import requests
import json
import pandas as pd
from datetime import datetime
```

```python
question_url = f"https://bbs-api.miyoushe.com/post/wapi/getPostFull"
```

```python



comments_url = f"https://bbs-api.miyoushe.com/post/wapi/getPostReplies"
```

```python
# 设置请求头
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
    "Accept": "application/json, text/plain, */*",
    "Origin": "https://www.miyoushe.com",
    "Referer": "https://www.miyoushe.com/"
}
```

```python
question_params = {
    "post_id": "60196460"
}
question_response = requests.get(question_url, params=question_params, headers=headers).json()

# 修改请求参数
comments_params = {
    "post_id": "60196460",  # 替换为实际文章ID
    "size": 20,
    "last_id": 0,
    "is_hot": "true"
}
```

```python
comments_response = requests.get(comments_url, params=comments_params).json()
```
