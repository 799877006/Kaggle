```python
import pandas as pd
import numpy as np
df = pd.read_csv('dataset/miyoushe_dataset.csv')
df2 = pd.read_csv('dataset/miyoushe_dataset2.csv')
df3 = pd.read_csv('dataset/miyoushe_dataset3.csv')
```

```python
character_list = ['那维莱特','恰斯卡','希诺宁','钟离','欧洛伦','纳西妲','胡桃','希诺宁','千织','基尼奇','雷电将军','玛拉妮','卡齐娜','艾梅莉埃','夜兰','妮露','娜维娅','希格雯','芙宁娜','克洛琳德','赛索斯','流浪者','白术','阿蕾齐诺','林尼','枫原万叶','荒泷一斗','优菈','八重神子','魈','闲云','嘉明','宵宫','夏沃蕾','神里绫华','赛诺','神里绫人','夏洛蒂','莱欧斯利','艾尔海森','温迪','达达利亚','琳妮特','珊瑚宫心海','可莉','申鹤','迪希雅','甘雨','提纳里','阿贝多','埃洛伊','七七','刻晴','迪卢克','莫娜','琴','旅行者','卡维','菲米尼','行秋','久歧忍','珐露珊','绮良良','五郎','瑶瑶','米卡','莱依拉','坎蒂丝','柯莱','多莉','鹿野院平藏','凝光','托马','九条裟罗','云堇','罗莎莉亚','早柚','雷泽','菲谢尔','班尼特','烟绯','重云','芭芭拉','迪奥娜','砂糖','诺艾尔','凯亚','辛焱','香菱','北斗','安伯','丽莎']
```

```python
# 创建一个函数来判断标题是否符合条件
def is_valid_title(title):
    return any(character in title for character in character_list) or "角色攻略" in title

# 分别从三个数据框中筛选符合条件的行
filtered_df = df[df['title'].apply(is_valid_title)]
filtered_df2 = df2[df2['title'].apply(is_valid_title)]
filtered_df3 = df3[df3['title'].apply(is_valid_title)]

# 合并三个筛选后的数据框
merged_df = pd.concat([filtered_df, filtered_df2, filtered_df3], ignore_index=True)

# 保存到新的CSV文件
merged_df.to_csv('dataset/merged_character_guides.csv', index=False)

# 显示合并后的数据框信息
print(f"合并后的数据总行数: {len(merged_df)}")
print("\n数据预览:")
print(merged_df.head())
```
